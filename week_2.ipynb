```python
#Calculating area
a=gpd.GeoSeries(df['geometry'])
df['area']=a.area
min_size=min(a.area)
std_dev=statistics.stdev(a.length)
# df
# min_size
max_size=max(a.area)
# max_size
avg_size=statistics.mean((a.area))
# avg_size
print(min_size)
print(max_size)
print(avg_size)
print(std_dev)
```
